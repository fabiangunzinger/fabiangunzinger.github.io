Helpers and setup

```python
def read(filepath, **kwargs):
    columns = ['SpendOut', 'TxnDate', 'UserCountry']
    return ha.s3read_csv(filepath, usecols=columns, **kwargs)

def select(df):
    mask = df.SpendOut.between(-1000, 0, inclusive='neither')
    return df.loc[mask]

def combine(pieces):
    base = pd.DataFrame()
    for piece in pieces:
        base = (base.append(piece)
                .pipe(agg))
    return base

def agg(df):
    return df.groupby(['TxnDate', 'UserCountry'], as_index=False).SpendOut.sum()


nrows = 1000
nfiles = 10
p = 'Main_Core_2021_02_\d\d'
filepaths = [''.join(['s3://', fp]) for fp in fs.ls('3di-data-fable/raw/Daily/Core') if re.search(p, fp)]
filepaths = filepaths[:nfiles]
```

Read sequentially

```python
%%time

pieces = []
for filepath in filepaths:
    df = read(filepath, nrows=nrows)
    df = select(df)
    pieces.append(df)

data = combine(pieces)
print(data.shape)
data.head()
```

Read in parallel

```python
%%time

pieces = []
for filepath in filepaths:
    df = delayed(read)(filepath, nrows=nrows)
    df = delayed(select)(df)
    pieces.append(df)

data = delayed(combine)(pieces)
```

```python
data.visualize()
```

```python
import time
for sch in ['threading', 'processes', 'sync']:
    t0 = time.time()
    r = data.compute(scheduler=sch)
    t1 = time.time()
    print(f"{sch:>10}, {t1 - t0:0.4f} s; result, {r.shape}")
```
